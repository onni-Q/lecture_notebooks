# Linear Regression

In a very simple linear regression prediction, we predict numerical values of one variable using the values of a second numerical variable. Here, we have 21 days of Facebook page views after you started a new campaign. We would like to predict how many views the computer expects.

This time we ask ChatGPT, and it delivers:

```python
#Keep cell
#https://chat.openai.com/share/4272d3e7-704c-4d5b-90da-b182421e46d2

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Given data
days = list(range(0, 21))
facebook = [5, 7, 4, 9, 11, 10, 14, 17, 13, 11, 18, 17, 21, 21, 24, 23, 28, 35, 21, 27, 23]

# Reshape the data to make it compatible with scikit-learn
days_reshaped = np.array(days).reshape(-1, 1)
facebook_reshaped = np.array(facebook)

# Create a linear regression model
model = LinearRegression()

# Fit the model to the data
model.fit(days_reshaped, facebook_reshaped)

# Make predictions
predictions = model.predict(days_reshaped)

# Plot the data and the linear regression line
plt.scatter(days, facebook, label='Actual data')
plt.plot(days, predictions, color='red', label='Linear Regression')
plt.xlabel('Days')
plt.ylabel('Facebook Likes')
plt.title('Linear Regression Model for Facebook Likes Over Days')
plt.legend()
plt.show()
```

Let's split this up into the steps involved.

First we load relevant libraries that you should already know about and create the data.

```python
#Keep cell

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Given data
days = list(range(0, 21))
facebook = [5, 7, 4, 9, 11, 10, 14, 17, 13, 11, 18, 17, 21, 21, 24, 23, 28, 35, 21, 27, 23]
```

Next, we import the linear regression model from the sklearn (https://scikit-learn.org/stable/), which is the package that contains a lot of machine-learning algorithms. We will talk a lot more about it.

```python
#Keep cell
from sklearn.linear_model import LinearRegression
```

Sklearn likes its data in the form of numpy arrays in two dimensions. This is what the next cell does.

```python
#Keep cell
facebook_reshaped = np.array(facebook)
days_reshaped = np.array(days).reshape(-1, 1)
```

We call facebook_reshaped the target and days_reshaped the input.

You might have noticed that the input has been 'reshaped' with `reshape`. Let's print out both `days` and `days_reshaped`:

```python
#Keep cell
print(days)
print(days_reshaped)
```

As you can hopefully see, sklearn expects the input to be not a list/array but as a matrix - in this case with only one column. This is to allow for more than only type of input later on. Just remember for now that if you have a simple list/array, you have to use `reshape(-1,1)`. For details, how this works, check out https://www.w3schools.com/python/numpy/numpy_array_reshape.asp and do practice a lot.

We move on and create the linear regression model using the instructions of ChatGPT.

```python
#Keep cell
# Create a linear regression model
model = LinearRegression()
```

Now, we 'fit' the model, which means we train it with the input data.

```python
#Keep cell
# Fit the model to the data
model.fit(days_reshaped, facebook_reshaped)
```

The next step is to make predictions.

```python
#Keep cell
# Make predictions
predictions = model.predict(days_reshaped)
```

Print out `predictions`.

```python
#Keep cell
predictions
```

These are the number of views per day the model predicts. Do you see the limitations of a linear model at the end? It cannot really fit well a decreasing trend, which is typical to campaigns. They become less interesting over time. A linear model cannot fit this.

Finally, we plot our results with the help of ChatGPT.

```python
#Keep cell
# Plot the data and the linear regression line
plt.scatter(days, facebook, label='Actual data')
plt.plot(days, predictions, color='red', label='Linear Regression')
plt.xlabel('Days')
plt.ylabel('Facebook Likes')
plt.title('Linear Regression Model for Facebook Likes Over Days')
plt.legend()
plt.show()
```
